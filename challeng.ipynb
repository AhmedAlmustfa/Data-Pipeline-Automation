```python
api_key = os.getenv('EIA_API_KEY')

api_path = "electricity/rto/region-sub-ba-data/data/"

frequency = "hourly"

facets = {
    "parent": "CISO",
    "subba": "SDGE"
}
```

```python
df1 = api.eia_get(
    api_key = api_key,
    api_path = api_path,
    frequency = frequency,
    facets = facets
)
```

```python
df1.parameters
```

```python

px.line(df1.data, x= "period", y= "value")
```

```python
start = datetime.datetime(2024, 1, 1, 1)
end = datetime.datetime(2024, 12, 31, 23)

df2 = api.eia_get(
    api_key = api_key,
    api_path = api_path,
    frequency = frequency,
    facets = facets,
    start = start,
    end = end
)
```
